```python
import pandas as pd
from sqlalchemy import create_engine
df = pd.read_csv("/content/country_vaccination_stats.csv")

# Create a connection to the SQL database
engine = create_engine('sqlite:///my_data1.db')

# Insert data into a SQL table
df.to_sql('vaccination_stats', engine, if_exists='replace', index=False)
```

```python
%sql SELECT * FROM vaccination_stats LIMIT 0
```

```python
%%sql
UPDATE vaccination_stats
SET daily_vaccinations = 
    CASE 
        WHEN daily_vaccinations IS NULL OR daily_vaccinations = 'None' OR daily_vaccinations = 'NaN' THEN 0 
        ELSE daily_vaccinations 
    END;

UPDATE vaccination_stats
SET daily_vaccinations = (
    SELECT DISTINCT MEDIAN(daily_vaccinations) OVER (PARTITION BY country)
    FROM vaccination_stats
    WHERE daily_vaccinations != 0 AND daily_vaccinations IS NOT NULL AND daily_vaccinations != 'None'
)
WHERE daily_vaccinations = 0;

UPDATE vaccination_stats
SET daily_vaccinations = (
    SELECT DISTINCT MEDIAN(daily_vaccinations) OVER (PARTITION BY country)
    FROM vaccination_stats
    WHERE country = vaccination_stats.country AND daily_vaccinations != 0 AND daily_vaccinations IS NOT NULL AND daily_vaccinations != 'None'
)
WHERE daily_vaccinations IS NULL OR daily_vaccinations = 'None' OR daily_vaccinations = 'NaN';
```
